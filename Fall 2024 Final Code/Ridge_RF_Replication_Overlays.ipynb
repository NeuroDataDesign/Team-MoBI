```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Sample_Fractions = [0.058823, 0.24252, 1]

# Import Raw Data
Data_posdif = pd.read_csv("y_posdif_23_trial_1_mean.csv")
Data_pos = pd.read_csv("y_pos_23_trial_1_mean.csv")
Data_speed = pd.read_csv("y_speed_23_trial_1_mean.csv")

Ridge_posdif = pd.read_csv('Ridge_Replication_posdif.csv')
Ridge_pos = pd.read_csv('Ridge_Replication_pos.csv')
Ridge_speed = pd.read_csv('Ridge_Replication_speed.csv')
```

```python
from matplotlib.ticker import FormatStrFormatter
fig, ax = plt.subplots(figsize=(8, 6))

# Position Difference
for i in range(5):
    ax.plot(Sample_Fractions, Data_posdif.iloc[i, :], color = 'orange', alpha=0.6, linewidth = 3, label='Position Difference' if 'Position Difference' not in ax.get_legend_handles_labels()[1] else "")
# Position
# Raw Data for Position Difference
for i in range(5):
    ax.plot(Sample_Fractions, Data_pos.iloc[i, :], color = 'blue', alpha=0.6, linewidth = 3, label='Stimulus Position' if 'Stimulus Position' not in ax.get_legend_handles_labels()[1] else "")
# Speed
for i in range(5):
    ax.plot(Sample_Fractions, Data_speed.iloc[i, :], color = 'green', alpha=0.6, linewidth = 3, label='Stimulus Speed' if 'Stimulus Speed' not in ax.get_legend_handles_labels()[1] else "")

# horizontal grid lines
ax.hlines(1, 0, 1, linestyle="--", color="black", alpha = 0.6)
ax.hlines(0.8, 0, 1, linestyle="--", color="red", alpha = 0.6)

ax.set_xlabel("Sample Fraction", fontsize=18)
ax.set_ylabel("R2 Score", fontsize=18)
ax.legend(frameon=True, fontsize=18, loc = 4)
ax.set_ylim(0, 1.1)
plt.tick_params(axis='x', labelsize=18)
plt.tick_params(axis='y', labelsize=18)
plt.title("Random Forest Replicated R2 Scores", fontsize=18)
plt.show()

fig, ax = plt.subplots(figsize=(8, 6))

# Position Difference
for i in range(5):
    ax.plot(Sample_Fractions, Ridge_posdif.iloc[i, :], color = 'orange', alpha=0.6, linewidth = 3, label='Position Difference' if 'Position Difference' not in ax.get_legend_handles_labels()[1] else "")
# Position
# Raw Data for Position Difference
for i in range(5):
    ax.plot(Sample_Fractions, Ridge_pos.iloc[i, :], color = 'blue', alpha=0.6, linewidth = 3, label='Stimulus Position' if 'Stimulus Position' not in ax.get_legend_handles_labels()[1] else "")
# Speed
for i in range(5):
    ax.plot(Sample_Fractions, Ridge_speed.iloc[i, :], color = 'green', alpha=0.6, linewidth = 3, label='Stimulus Speed' if 'Stimulus Speed' not in ax.get_legend_handles_labels()[1] else "")

ax.set_xlabel("Sample Fraction", fontsize=18)
ax.set_ylabel("R2 Score", fontsize=18)
ax.legend(frameon=True, fontsize=18, loc = 4)
plt.tick_params(axis='x', labelsize=18)
plt.tick_params(axis='y', labelsize=18)
ax.xaxis.set_major_formatter(FormatStrFormatter('%.2f'))  # Limit to 2 decimal places
ax.yaxis.set_major_formatter(FormatStrFormatter('%.2f'))
plt.title("Ridge Replicated R2 Scores", fontsize=18)
plt.show()
```

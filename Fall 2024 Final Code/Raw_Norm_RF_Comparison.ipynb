# Normalization Overlays

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Sample_Fractions = [0.058823, 0.24252, 1]

# Import Raw Data
Data_posdif = pd.read_csv("All Normalized and Raw (5subj) CSVs/y_pos_dif_r2_scores_23_trial_1_5subjects_mean.csv")
Data_pos = pd.read_csv("All Normalized and Raw (5subj) CSVs/y_pos_r2_scores_23_trial_1_5subjects_mean.csv")
Data_speed = pd.read_csv("All Normalized and Raw (5subj) CSVs/y_speed_r2_scores_23_trial_1_5subjects_mean.csv")

# Import Min-Max Data
MMData_posdif = pd.read_csv("All Normalized and Raw (5subj) CSVs/MM_y_pos_dif_r2_scores_23_trial_1_5subjects_mean.csv")
MMData_pos = pd.read_csv("All Normalized and Raw (5subj) CSVs/MM_y_pos_r2_scores_23_trial_1_5subjects_mean.csv")
MMData_speed = pd.read_csv("All Normalized and Raw (5subj) CSVs/MM_y_speed_r2_scores_23_trial_1_5subjects_mean.csv")

# Import Z-Score Data
ZSData_posdif = pd.read_csv("All Normalized and Raw (5subj) CSVs/ZS_y_pos_dif_5_subjects_mean.csv")
ZSData_pos = pd.read_csv("All Normalized and Raw (5subj) CSVs/ZS_y_pos_5_subjects_mean.csv")
ZSData_speed = pd.read_csv("All Normalized and Raw (5subj) CSVs/ZS_y_speed_5_subjects_mean.csv")

# Import Robust Scaling Data
RSData_posdif = pd.read_csv("All Normalized and Raw (5subj) CSVs/RS_y_pos_dif_r2_scores_23_trial_1_5subjects_mean.csv")
RSData_pos = pd.read_csv("All Normalized and Raw (5subj) CSVs/RS_y_pos_r2_scores_23_trial_1_5subjects_mean.csv")
RSData_speed = pd.read_csv("All Normalized and Raw (5subj) CSVs/RS_y_speed_r2_scores_23_trial_1_5subjects_mean.csv")
```

## Plotting

### For All targets

```python
fig, ax = plt.subplots(figsize=(8, 6))

# Raw Data for Position Difference
for i in range(1):
    ax.plot(Sample_Fractions, Data_posdif.iloc[i, :], color = 'blue', alpha=0.6, linewidth = 5, label='Raw Data' if 'Raw Data' not in ax.get_legend_handles_labels()[1] else "")
# MM Norm Data for Position Difference
for i in range(1):
    ax.plot(Sample_Fractions, MMData_posdif.iloc[i, :], color='yellow', linewidth = 4, linestyle = '-.', label='Min-Max Normalization' if 'Min-Max Normalization' not in ax.get_legend_handles_labels()[1] else "")
# ZS Norm Data for Position Difference
for i in range(1):
    ax.plot(Sample_Fractions, ZSData_posdif.iloc[i, :], color='green', linewidth = 3.5, linestyle = '--', label='Z-Score Normalization' if 'Z-Score Normalization' not in ax.get_legend_handles_labels()[1] else "")
# ZS Norm Data for Position Difference
for i in range(1):
    ax.plot(Sample_Fractions, RSData_posdif.iloc[i, :], color='red', linewidth = 3, linestyle = ':', label='Robust Scaling Normalization' if 'Robust Scaling Normalization' not in ax.get_legend_handles_labels()[1] else "")

# horizontal grid lines
ax.hlines(1, 0, 1, linestyle="--", color="black", alpha = 0.6)
ax.hlines(0.8, 0, 1, linestyle="--", color="red", alpha = 0.6)

ax.set_xlabel("Sample Fraction", fontsize=18)
ax.set_ylabel("R2 Score", fontsize=18)
ax.legend(frameon=False, fontsize=18, loc=4)
ax.set_ylim(0,None)
plt.tick_params(axis='x', labelsize=18)
plt.tick_params(axis='y', labelsize=18)
plt.title("Position Difference Random Forest R2 Scores", fontsize=18)
plt.show()

fig, ax = plt.subplots(figsize=(8, 6))

# Raw Data for Position Difference
for i in range(1):
    ax.plot(Sample_Fractions, Data_pos.iloc[i, :], color = 'blue', alpha=0.6, linewidth = 5, label='Raw Data' if 'Raw Data' not in ax.get_legend_handles_labels()[1] else "")
# MM Norm Data for Position Difference
for i in range(1):
    ax.plot(Sample_Fractions, MMData_pos.iloc[i, :], color='yellow', linewidth = 4, linestyle = '-.', label='Min-Max Normalization' if 'Min-Max Normalization' not in ax.get_legend_handles_labels()[1] else "")
# ZS Norm Data for Position Difference
for i in range(1):
    ax.plot(Sample_Fractions, ZSData_pos.iloc[i, :], color='green', linewidth = 3.5, linestyle = '--', label='Z-Score Normalization' if 'Z-Score Normalization' not in ax.get_legend_handles_labels()[1] else "")
# ZS Norm Data for Position Difference
for i in range(1):
    ax.plot(Sample_Fractions, RSData_pos.iloc[i, :], color='red', linewidth = 3, linestyle = ':', label='Robust Scaling Normalization' if 'Robust Scaling Normalization' not in ax.get_legend_handles_labels()[1] else "")

# horizontal grid lines
ax.hlines(1, 0, 1, linestyle="--", color="black", alpha = 0.6)
ax.hlines(0.8, 0, 1, linestyle="--", color="red", alpha = 0.6)

ax.set_xlabel("Sample Fraction", fontsize=18)
ax.set_ylabel("R2 Score", fontsize=18)
ax.legend(frameon=False, fontsize=18, loc=4)
ax.set_ylim(0,None)
plt.tick_params(axis='x', labelsize=18)
plt.tick_params(axis='y', labelsize=18)
plt.title("Stimulus Position Random Forest R2 Scores", fontsize=18)
plt.show()

fig, ax = plt.subplots(figsize=(8, 6))

# Raw Data for Position Difference
for i in range(1):
    ax.plot(Sample_Fractions, Data_speed.iloc[i, :], color = 'blue', alpha=0.6, linewidth = 5, label='Raw Data' if 'Raw Data' not in ax.get_legend_handles_labels()[1] else "")
# MM Norm Data for Position Difference
for i in range(1):
    ax.plot(Sample_Fractions, MMData_speed.iloc[i, :], color='yellow', linewidth = 4, linestyle = '-.', label='Min-Max Normalization' if 'Min-Max Normalization' not in ax.get_legend_handles_labels()[1] else "")
# ZS Norm Data for Position Difference
for i in range(1):
    ax.plot(Sample_Fractions, ZSData_speed.iloc[i, :], color='green', linewidth = 3.5, linestyle = '--', label='Z-Score Normalization' if 'Z-Score Normalization' not in ax.get_legend_handles_labels()[1] else "")
# ZS Norm Data for Position Difference
for i in range(1):
    ax.plot(Sample_Fractions, RSData_speed.iloc[i, :], color='red', linewidth = 3, linestyle = ':', label='Robust Scaling Normalization' if 'Robust Scaling Normalization' not in ax.get_legend_handles_labels()[1] else "")

# horizontal grid lines
ax.hlines(1, 0.2, 1, linestyle="--", color="black", alpha = 0.6)
ax.hlines(0.8, 0.2, 1, linestyle="--", color="red", alpha = 0.6)

ax.set_xlabel("Sample Fraction", fontsize=18)
ax.set_ylabel("R2 Score", fontsize=18)
ax.legend(frameon=False, fontsize=18, loc = 4)
ax.set_ylim(0,1.05)
plt.xlim(0.15, None)
plt.tick_params(axis='x', labelsize=18)
plt.tick_params(axis='y', labelsize=18)
plt.title("Stimulus Speed Random Forest R2 Scores", fontsize=18)
plt.show()
```

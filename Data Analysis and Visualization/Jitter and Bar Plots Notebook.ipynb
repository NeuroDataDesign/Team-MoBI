# Jitter and Bar Plot Notebook

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd

Sample_Fractions = [0.058823, 0.24252, 1]
People = ['Isha', 'Avva', 'Liora', 'Deborah', 'Taiming', 'Riya']
Pos_Dif_R2_Values = [
    [-1.413213396,	0.891728281,	0.985842446],
    [-1.413213396,	0.891728281,	0.985842446],
    [-1.413213396,	0.891728281,	0.985842446],
    [-1.413213396,	0.891728281,	0.985842446],
    [-1.413213396,	0.891728281,	0.985842446],
    [-1.413213396,	0.891728281,	0.985842446]
]

posdif_data = []
for i, person in enumerate(People):
    for j, fraction in enumerate(Sample_Fractions):
        posdif_data.append({'Person': person, 'Sample_Fraction': fraction, 'R2_Value': Pos_Dif_R2_Values[i][j]})

posdif_df = pd.DataFrame(posdif_data)

Pos_R2_Values = [
    [-0.720993759,	0.934434798,	0.992364234],
    [-0.720993759,	0.934434798,	0.992364234],
    [-0.720993759,	0.934434798,	0.992364234],
    [-0.720993759,	0.934434798,	0.992364234],
    [-0.720993759,	0.934434798,	0.992364234],
    [-0.720993759,	0.934434798,	0.992364234]
]

pos_data = []
for i, person in enumerate(People):
    for j, fraction in enumerate(Sample_Fractions):
        pos_data.append({'Person': person, 'Sample_Fraction': fraction, 'R2_Value': Pos_R2_Values[i][j]})

pos_df = pd.DataFrame(pos_data)

Speed_R2_Values = [
    [-4.693400898,	0.232379264,	0.951094028],
    [-4.693400898,	0.232379264,	0.951094028],
    [-4.693400898,	0.232379264,	0.951094028],
    [-4.693400898,	0.232379264,	0.951094028],
    [-4.693400898,	0.232379264,	0.951094028],
    [-4.693400898,	0.232379264,	0.951094028]
]

speed_data = []
for i, person in enumerate(People):
    for j, fraction in enumerate(Sample_Fractions):
        speed_data.append({'Person': person, 'Sample_Fraction': fraction, 'R2_Value': Speed_R2_Values[i][j]})

speed_df = pd.DataFrame(speed_data)

```

```python
plt.figure(figsize=(6, 3.5))
sns.stripplot(data=posdif_df, x="Sample_Fraction", y="R2_Value", hue="Person", jitter=True, palette="Reds", dodge=True)
plt.suptitle("RF Regression Mean R Values for All Team Members")
plt.title("Position Difference - One Subject")
plt.xlabel("Sample Fraction")
plt.ylabel("R Value")
plt.legend(title="Person")
plt.show()

plt.figure(figsize=(6, 3.5))
sns.stripplot(data=pos_df, x="Sample_Fraction", y="R2_Value", hue="Person", jitter=True, palette="Reds", dodge=True)
plt.suptitle("RF Regression Mean R Values for All Team Members")
plt.title("Stimulus Position - One Subject")
plt.xlabel("Sample Fraction")
plt.ylabel("R Value")
plt.legend(title="Person")
plt.show()

plt.figure(figsize=(6, 3.5))
sns.stripplot(data=speed_df, x="Sample_Fraction", y="R2_Value", hue="Person", jitter=True, palette="Reds", dodge=True)
plt.suptitle("RF Regression Mean R Values for All Team Members")
plt.title("Stimulus Speed - One Subject")
plt.xlabel("Sample Fraction")
plt.ylabel("R Value")
plt.legend(title="Person")
plt.show()
```

```python
Pos_Dif_R2_Values = np.array(Pos_Dif_R2_Values)
Pos_R2_Values = np.array(Pos_R2_Values)
Speed_R2_Values = np.array(Speed_R2_Values)

num_fractions = len(Sample_Fractions)
num_individuals = len(People)

x = np.arange(num_fractions)

# Plotting the grouped bar plot
fig, ax = plt.subplots()
for i in range(num_individuals):
    # Position each bar group for the individual
    ax.bar(x + i * 0.15, Pos_Dif_R2_Values[i], width=0.15, label=People[i])

# Add labels, title, and legend
ax.set_xlabel('Sample Fractions')
ax.set_ylabel('R Values')
plt.title('RF Regression for |user_pos - stim_pos|')
plt.suptitle('One Subject for All Systems')
ax.set_xticks(x + 0.15 * (num_individuals - 1) / 2)
ax.set_xticklabels(Sample_Fractions)
ax.set_ylim(-1, 1)
ax.legend(title="People")

plt.show()

# Plotting the grouped bar plot
fig, ax = plt.subplots()
for i in range(num_individuals):
    # Position each bar group for the individual
    ax.bar(x + i * 0.15, Pos_R2_Values[i], width=0.15, label=People[i])

# Add labels, title, and legend
ax.set_xlabel('Sample Fractions')
ax.set_ylabel('R Values')
plt.title('RF Regression for |stim_pos|')
plt.suptitle('One Subject for All Systems')
ax.set_xticks(x + 0.15 * (num_individuals - 1) / 2)
ax.set_xticklabels(Sample_Fractions)
ax.legend(title="People")

plt.show()

# Plotting the grouped bar plot
fig, ax = plt.subplots()
for i in range(num_individuals):
    # Position each bar group for the individual
    ax.bar(x + i * 0.15, Speed_R2_Values[i], width=0.15, label=People[i])

# Add labels, title, and legend
ax.set_xlabel('Sample Fractions')
ax.set_ylabel('R Values')
plt.title('RF Regression for |change_rate_x|')
plt.suptitle('One Subject for All Systems')
ax.set_xticks(x + 0.15 * (num_individuals - 1) / 2)
ax.set_xticklabels(Sample_Fractions)
ax.set_ylim(-1, 1)
ax.legend(title="People")

plt.show()
```

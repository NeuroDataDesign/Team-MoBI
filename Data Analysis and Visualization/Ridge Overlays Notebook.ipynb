# Manual and Sklearn Ridge Regression Overlays

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import data for Manual Ridge
manual_data = pd.read_csv('ridge_regression_raw_r2_scores.csv')
sklearn_data = pd.read_csv('sklearn_ridge_regression_raw_r2_scores.csv')
```

### Plotting

```python
manual_lambda_data = manual_data[(manual_data['Lambda'] == 0.1) & (manual_data['Subject'].isin([f'Subject {i}' for i in range(1, 6)]))]
sklearn_lambda_data = sklearn_data[(sklearn_data['Lambda'] == 0.1) & (sklearn_data['Subject'].isin([f'Subject {i}' for i in range(1, 6)]))]

fig, ax = plt.subplots(figsize=(10, 6))

for subject in manual_lambda_data['Subject'].unique():
    subject_data = manual_lambda_data[manual_lambda_data['Subject'] == subject]
    ax.plot(subject_data['Sample Fraction'], subject_data['R^2 Score'], color = 'yellow', linewidth = 3, label= 'Manual Ridge' if 'Manual Ridge' not in ax.get_legend_handles_labels()[1] else "" )

for subject in sklearn_lambda_data['Subject'].unique():
    subject_data = sklearn_lambda_data[sklearn_lambda_data['Subject'] == subject]
    ax.plot(subject_data['Sample Fraction'], subject_data['R^2 Score'], color = 'blue', linewidth = 3, linestyle = '--', label= 'Sklearn Ridge' if 'Sklearn Ridge' not in ax.get_legend_handles_labels()[1] else "" )

ax.set_ylim(0,None)
plt.xlabel('Sample Fraction')
plt.ylabel('R Score')
plt.title('R Scores vs Sample Fraction (Lambda = 0.1, First 5 Subjects)')
plt.legend()
plt.show()
```

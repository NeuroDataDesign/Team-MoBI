# Normalization Overlays

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Sample_Fractions = [0.058823, 0.24252, 1]

# Import Raw Data
Data_posdif = pd.read_csv('y_pos_dif_r2_scores_23_trial_1_5subjects_mean.csv')
Data_pos = pd.read_csv('y_pos_r2_scores_23_trial_1_5subjects_mean.csv')
Data_speed = pd.read_csv('y_speed_r2_scores_23_trial_1_5subjects_mean.csv')

# Import Min-Max Data
MMData_posdif = pd.read_csv('MM_y_pos_dif_r2_scores_23_trial_1_5subjects_mean.csv')
MMData_pos = pd.read_csv('MM_y_pos_r2_scores_23_trial_1_5subjects_mean.csv')
MMData_speed = pd.read_csv('MM_y_speed_r2_scores_23_trial_1_5subjects_mean.csv')

# Import Z-Score Data
ZSData_posdif = pd.read_csv('ZS_y_pos_dif_5_subjects_mean.csv')
ZSData_pos = pd.read_csv('ZS_y_pos_5_subjects_mean.csv')
ZSData_speed = pd.read_csv('ZS_y_speed_5_subjects_mean.csv')

# Import Robust Scaling Data
RSData_posdif = pd.read_csv('RS_y_pos_dif_r2_scores_23_trial_1_5subjects_mean.csv')
RSData_pos = pd.read_csv('RS_y_pos_r2_scores_23_trial_1_5subjects_mean.csv')
RSData_speed = pd.read_csv('RS_y_speed_r2_scores_23_trial_1_5subjects_mean.csv')
```

## Plotting

### For Position Difference

```python
fig, ax = plt.subplots(figsize=(8, 7))

# Raw Data for Position Difference
for i in range(5):
    ax.plot(Sample_Fractions, Data_posdif.iloc[i, :], color = 'red', alpha=0.7, linewidth = 4.5, label='Raw Data' if 'Raw Data' not in ax.get_legend_handles_labels()[1] else "")
# MM Norm Data for Position Difference
for i in range(5):
    ax.plot(Sample_Fractions, MMData_posdif.iloc[i, :], color='blue', alpha=0.6, linewidth = 3, linestyle = '--', label='Min-Max Normalized Data' if 'Min-Max Normalized Data' not in ax.get_legend_handles_labels()[1] else "")
    # ZS Norm Data for Position Difference
for i in range(5):
    ax.plot(Sample_Fractions, ZSData_posdif.iloc[i, :], color='green', alpha=0.8, linestyle='-.', linewidth = 2, label='Z-Score Normalized Data' if 'Z-Score Normalized Data' not in ax.get_legend_handles_labels()[1] else "")
# RS Norm Data for Position Difference
for i in range(5):
    ax.plot(Sample_Fractions, RSData_posdif.iloc[i, :], color='yellow', alpha=1, linestyle=':', linewidth = 1.5, label='Robust Scaling Normalized Data' if 'Robust Scaling Normalized Data' not in ax.get_legend_handles_labels()[1] else "")

# horizontal grid lines
ax.hlines(1, 0, 1, linestyle="--", color="black", label="Max", alpha = 0.6)
ax.hlines(0.8, 0, 1, linestyle="--", color="red", label="0.8", alpha = 0.6)

ax.set_xlabel("Sample Fraction", fontsize=15)
ax.set_ylabel("R2 Score", fontsize=15)
ax.legend(frameon=False, fontsize=12, loc=4)
ax.set_ylim(0.7,1.01)
ax.set_xlim(0.2,1.01)
plt.title("RF Regression for |user_pos - stim_pos|: First 5 Subjects", fontsize=15)
plt.show()
```
